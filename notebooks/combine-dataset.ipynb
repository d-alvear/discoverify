```python
# collected.drop('top_genre',axis=1,inplace=True)

filtered = collected[collected['subgenres'].isnull()==True]

for i,artist_id in enumerate(filtered['artist_id'].unique()):
    l = len(filtered['artist_id'].unique())
    
    search = sp.artist(artist_id)
    genre_list = search['genres']
    
    artist_genre[artist_id] = str(genre_list)
        
    clear_output(wait=True)
    print(f"{i}/{l}")
        
artist_genre
```

```python
collected['subgenres'] = collected['artist_id'].map(artist_genre)
```

```python
from ast import literal_eval

def get_top_subgenre(list_string):
    try:
        lst = literal_eval(list_string)
    
        return lst[0]
    except:
        pass

collected['top_subgenre'] = collected['subgenres'].apply(lambda x: get_top_subgenre(x))

collected
```
