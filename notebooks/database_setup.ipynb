```python
import pandas as pd
import numpy as np

import json

import psycopg2 as pg
conn = pg.connect(database="spotify_db",
                  user="postgres", 
                  password="damara1004")

def run_query(q):
    with conn:
        try:
            cur = conn.cursor()
            cur.execute(q)
            return pd.read_sql(q, conn)

        except (Exception, pg.DatabaseError) as error:
            print(error)

def run_command(c):
    with conn:
        try:
            cur = conn.cursor()
            cur.execute(c)
            cur.close()
            conn.commit()
            
        except (Exception, pg.DatabaseError) as error:
            print(error)
```
